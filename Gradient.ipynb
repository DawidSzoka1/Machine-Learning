```python
import numpy as np
import pandas as pd
import plotly.express as px

np.random.seed(42)
```

```python
X1 = np.array([1, 2, 3, 4, 5, 6])
Y = np.array([3000, 3250, 3500, 3750, 4000, 4250])
m = len(X1)
```

```python
X1 = X1.reshape(m, 1)
Y = Y.reshape(-1, 1)
print(X1)
print(Y)
```

```python
bias = np.ones((m,1))
print(bias)
```

```python
X = np.append(bias, X1, axis=1)
print(X)
print(X.shape)
```

```python
learning_rate = 0.01

W = np.random.randn(2,1)
print(X)
print(W)
```

```python
intercept = []
coef = []

for i in range(3000):
  gradient = (2/m) * X.T.dot(X.dot(W) - Y)
  W = W - learning_rate * gradient
  intercept.append(W[0][0])
  coef.append(W[1][0])

print(W)
```

```python
df = pd.DataFrame(data={'intercept': intercept, 'coef': coef})
```

```python
df.head()
```

```python
px.line(df, y='intercept', width=800)
```

```python
px.line(df, y='coef', width=800)
```
