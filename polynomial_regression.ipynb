```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import sklearn

np.random.seed(42)
np.set_printoptions(precision=6, suppress=True)
sns.set(font_scale=1.3)
sklearn.__version__
```

```python
X = np.arange(-10, 10, 0.5)

noise = 80 * np.random.randn(40)
y = -X ** 3 + 10 * X ** 2 - 2 * X + 3 + noise
X = X.reshape(40, 1)

plt.figure(figsize=(8, 6))
plt.title('title')
plt.xlabel('X')
plt.ylabel('y')
plt.scatter(X, y, label='X')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X, y)

y_pred_lin = regressor.predict(X)

plt.figure(figsize=(8, 6))
plt.title("liniowa")
plt.xlabel('X')
plt.ylabel('y')
plt.scatter(X, y, label='X')
plt.plot(X, y_pred_lin, c='red', label='liniowa')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score

r2_score(y, y_pred_lin)
```

```python
df = pd.DataFrame(data={'X': X.ravel()})
df.head(10)
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2)

X_poly = poly.fit_transform(X)
X_poly
```

```python
X_poly.shape
```

```python
regressor = LinearRegression()
regressor.fit(X_poly, y)

y_pred_two = regressor.predict(X_poly)
```

```python
plt.figure(figsize=(8, 6))
plt.xlabel('X')
plt.ylabel('y')
plt.scatter(X, y, label='X')
plt.plot(X, y_pred_two, c='green')
plt.legend()
plt.show()
```

```python
r2_score(y, y_pred_two)
```

```python
poly = PolynomialFeatures(degree=3)

X_poly_4 = poly.fit_transform(X)

regressor = LinearRegression()
regressor.fit(X_poly_4, y)

y_pred_4 = regressor.predict(X_poly_4)

plt.figure(figsize=(8, 6))
plt.xlabel('X')
plt.ylabel('y')
plt.scatter(X, y, label='X')
plt.plot(X, y_pred_4, label='4', color='red')
plt.legend()
plt.show()
```

```python
r2_score(y, y_pred_4)
```
